```python
import cv2 
vidcap = cv2.VideoCapture('/gdrive/My Drive/testvideo/test8.mp4') 
count = 0

success = True
while success: 
  success, image = vidcap.read() 
  image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
  
  # 30프레임당 하나씩 이미지 추출 
  if(int(vidcap.get(1)) % 1 == 0): 
    print('Saved frame number : ' + str(int(vidcap.get(1)))) 
    
    # 추출된 이미지가 저장되는 경로 
    cv2.imwrite("/gdrive/My Drive/frame3/%d.png" % count, image)
    print('Saved frame%d.jpg' % count)
    count += 1
```
